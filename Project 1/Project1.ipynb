```python
#let's plot the first graph, which is a histogram of 'Ladder Score', this gives us a general distribution of
#'happiness' per country.
plt.hist(df['Ladder score'], bins = 100)
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.title('Histogram of Ladder score')
plt.savefig('histogram_ladder_score.png')
plt.show()
```

This shows us that the highest happiness score falls a little below 8, and the lowest ones are a little below one. But the highest distributions fall at a little after 4 and 6.

```python
#Let's see a scatterplot of GDP per capita and happiness scores
sns.scatterplot(data=df, x='Logged GDP per capita', y='Ladder score')
plt.xlabel('Logged GDP per capita')
plt.ylabel('Happiness Score')
plt.title('Scatterplot of Logged GDP per capita vs Ladder score')
plt.savefig('scatterplot_gdp_vs_happiness.png')
plt.show()
```

The analysis shows strong correlations between happiness scores and GDP per capita, in fact, the higher the GDP, the higher the happiness score.

```python
#Let's generate a heatmap to determine the true correlation between values.
cols_for_heatmap = ['Ladder score', 'Logged GDP per capita', 'Social support',
                   'Healthy life expectancy', 'Freedom to make life choices',
                   'Generosity', 'Perceptions of corruption']
corr_matrix = df[cols_for_heatmap].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            fmt=".2f",
            linewidths=.5,
            annot_kws={"size": 10})


plt.title('Correlation Heatmap of Happiness Factors')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('easy_to_read_heatmap.png')
plt.show()
```

As shown in the heatmap, there are correlation between some variables, mainly GDP per capita, freedom to make life choices, healthy life expectancy,  and social support. There isn't a lot of it when compared to generosity or perceptions of corruption.

```python
#Let's see how we can compare happiness scores for the top and lower ten countries, and see how it relates regionally.
# Visualization 1: Distribution of Happiness Scores
plt.figure(figsize=(7, 6))
sns.histplot(df['Ladder score'], bins=10, kde=True, color='blue')
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Count')
plt.savefig('distribution_happiness_scores.png')

# Bar plot for a top ten countries
plt.figure(figsize=(12, 8))
sns.barplot(x='Ladder score', y='Country name', data=df.sort_values('Ladder score', ascending=False).head(10))
plt.title('Top 10 Countries by Happiness Score')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.savefig('top_10_countries.png')

# Bar plot for a lower ten countries
plt.figure(figsize=(12, 8))
sns.barplot(x='Ladder score', y='Country name', data=df.sort_values('Ladder score', ascending=True).head(10))
plt.title('Lower 10 Countries by Happiness Score')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.savefig('lower_10_countries.png') # Changed the save path



# Visualization 2: Regional GDP per Capita vs Ladder Score
plt.figure(figsize=(7, 6))
sns.scatterplot(x='Logged GDP per capita', y='Ladder score', hue='Regional indicator', data=df)
plt.title('GDP per Capita vs Happiness')
plt.xlabel('Logged GDP per Capita')
plt.ylabel('Happiness Score')
plt.savefig('gdp_vs_happiness.png')

# Visualization 3: Regional Social Support vs Ladder Score
plt.figure(figsize=(7, 6))
sns.scatterplot(x='Social support', y='Ladder score', hue='Regional indicator', data=df)
plt.title('Social Support vs Happiness')
plt.xlabel('Social Support')
plt.ylabel('Happiness Score')
plt.savefig('social_support_vs_happiness.png') # Changed the save path




graphs_paths = [
    'distribution_happiness_scores.png',
    'gdp_vs_happiness.png',
    'social_support_vs_happiness.png',
]
graphs_paths
```

From the distribution of happiness scores globally, we see that we have a normal distribution, with very few outliers, and the majority of the distribution of happiness falls between 4.5 and 6.5, which is pretty optimistic.

The top ten countries are listed, with Finland rankingg at number one on the happiness scale and the lower ten indicate that Afghanistan ranks the lowest.

Regionally, Western Europe ranks the highest in terms of GDP per capita and happiness, while sub-saharan africa ranks the lowest. For an additional regional analysis, we have social support compared to happiness score, and the same results are prevalent, with a few outliers here and there.


```python
#Now let's evaluate the happiness score of the top ten and the lower ten countries and see how they compare to
#one another:

top_10 = df.sort_values(by='Ladder score', ascending=False).head(10)
top_10_countries = top_10[['Country name', 'Logged GDP per capita', 'Ladder score']]


fig, ax1 = plt.subplots(figsize=(10, 6))


sns.barplot(x='Country name', y='Logged GDP per capita', data=top_10_countries, color='blue', ax=ax1)
ax1.set_ylabel('Logged GDP per capita', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')


ax2 = ax1.twinx()
sns.lineplot(x='Country name', y='Ladder score', data=top_10_countries, color='red', marker="o", ax=ax2)
ax2.set_ylabel('Happiness (Ladder Score)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.title('Comparison of GDP and Happiness for Top 10 Happiest Countries')

plt.savefig('gdp_happiness_comparison_top_ten.png')

plt.show()


top_10 = df.sort_values(by='Ladder score', ascending=True).head(10)
top_10_countries = top_10[['Country name', 'Logged GDP per capita', 'Ladder score']]


fig, ax1 = plt.subplots(figsize=(10, 6))


sns.barplot(x='Country name', y='Logged GDP per capita', data=top_10_countries, color='blue', ax=ax1)
ax1.set_ylabel('Logged GDP per capita', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')


ax2 = ax1.twinx()
sns.lineplot(x='Country name', y='Ladder score', data=top_10_countries, color='red', marker="o", ax=ax2)
ax2.set_ylabel('Happiness (Ladder Score)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.title('Comparison of GDP and Happiness for Lower 10 Happiest Countries')

plt.savefig('gdp_happiness_comparison_lower_ten.png')

plt.show()


```

The two graphs illustrate where the top countries are in terms of happiness compared to the bottom ten. The more GDP rises, the happier a country is. Money does fix a lot of issues, but are there other factors that would hinder or add to happiness?

I asked the data what it can tell me about a country's corruption and whether that could affect a country's happiness.

```python
country_corruption = df.groupby('Country name')['Perceptions of corruption'].mean().reset_index()
```

```python


# 1. Now let's analyze by corruption
top_10_corrupted = country_corruption.sort_values(by='Perceptions of corruption', ascending=False).head(10)

top_10_corrupted_with_happiness = pd.merge(top_10_corrupted, df[['Country name', 'Ladder score']], on='Country name')

plt.figure(figsize=(12, 6))
sns.barplot(x='Country name', y='Ladder score', data=top_10_corrupted_with_happiness, color='skyblue')
plt.title('Happiness Scores of Top 10 Most Corrupted Countries')
plt.xlabel('Country')
plt.ylabel('Happiness Score (Ladder score)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('happiness_vs_corruption.png')
plt.show()
```

This does not tell me much. The countries that ranked as the ten most corrupted countries are listed here, and even though Afghanistan appears as the bottom least happy, and also appears in the corruption chart, one can easily assume that more corruption means less happiness, but this chart tells us something different. Kosovo for example, has a higher than 6 happiness score, higher than the average, even though it ranks fifth as the most corrupt.

What about social support? Does a country's attitude towards social support also affects happiness?

```python

#What about social support?
least_social_support = df.sort_values(by='Social support', ascending=True).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Country name', y='Ladder score', data=least_social_support, color='salmon')
plt.title('Happiness Scores of Top 10 Countries with Least Social Support')
plt.xlabel('Country')
plt.ylabel('Happiness Score (Ladder score)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('happiness_vs_social_support.png')
plt.show()

plt.savefig('least_social_support_table.png')
print(least_social_support[['Country name', 'Social support', 'Ladder score']])
```

Now we can sort of see a true correlation as another factor emerges itself as having an impact on happiness score. Since the average happiness score is 5.47, and none of those countries listed score it, we can see the impact that lack of social support has in creating happiness in citizens of these countries.

Now that we have another factor at play, how does it compare to the top ten countries with the most social support?

```python
least_social_support = df.sort_values(by='Social support', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Country name', y='Ladder score', data=least_social_support, color='purple')
plt.title('Happiness Scores of Top 10 Countries with Most Social Support')
plt.xlabel('Country')
plt.ylabel('Happiness Score (Ladder score)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('happiness_vs_social_support.png')
plt.show()

plt.savefig('most_social_support_table.png')
print(least_social_support[['Country name', 'Social support', 'Ladder score']])
```

This confirms our thought process. The countries with the highest social support also score the highest on the happiness score.

Now let's compare freedom to make life choices for countries that rank the lowest compared to the average.

```python

#Let's compare healthy life expectancy
least_social_support = df.sort_values(by='Healthy life expectancy', ascending=True).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Country name', y='Ladder score', data=least_social_support, color='green')
plt.title('Happiness Scores of Top 10 Countries with Least life expectancy')
plt.xlabel('Country')
plt.ylabel('Happiness Score (Ladder score)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('happiness_vs_health_low.png')
plt.show()

plt.savefig('least_life_expectancy.png')
print(least_social_support[['Country name', 'Healthy life expectancy', 'Ladder score']])

```

all countries listed score low on the ladder score and also have citizens susceptible to not very healthy standards of living, so there exists a correlation between healthy life expectancy and happiness.

```python

#What about social support?
least_social_support = df.sort_values(by='Healthy life expectancy', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Country name', y='Ladder score', data=least_social_support, color='salmon')
plt.title('Happiness Scores of Top 10 Countries with Most Life Expectancy')
plt.xlabel('Country')
plt.ylabel('Happiness Score (Ladder score)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('happiness_vs_social_support.png')
plt.show()

plt.savefig('most_health.png')
print(least_social_support[['Country name', 'Healthy life expectancy', 'Ladder score']])
```

We clearly see a correlation between higher ranking happines score countries and a healthy life expectancy. All the countries listed above ranked higher than the average 5.4 happiness score.

```python
least_freedom = df.sort_values(by='Freedom to make life choices', ascending=True).head(10)

avg_happiness_least_freedom = least_freedom['Ladder score'].mean()

overall_avg_happiness = df['Ladder score'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=['Least Freedom', 'Overall Average'], y=[avg_happiness_least_freedom, overall_avg_happiness], color='lightgreen')
plt.title('Happiness Comparison: Least Freedom vs. Overall Average')
plt.ylabel('Happiness Score (Ladder score)')
plt.tight_layout()
plt.savefig('happiness_comparison_least_freedom.png')
plt.show()

print(f"Average Happiness Score (Least Freedom): {avg_happiness_least_freedom:.2f}")
print(f"Overall Average Happiness Score: {overall_avg_happiness:.2f}")
```

Again we see that for those countries that rank the lowest in freedom, their average happiness score also ranked the lowest, at 4.16, as opposed to the overall average countries with freedom to make life choices, whose happiness index is at 5.47

```python
# Countries with the least freedom
for i, country in enumerate(least_freedom['Country name']):
    print(f"{i+1}. {country}")
```

```python
#Now let's compare overall happiness of regional regions.


regional_happiness = df.groupby('Regional indicator')['Ladder score'].mean().reset_index()

regional_happiness = regional_happiness.sort_values(by='Ladder score', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Regional indicator', y='Ladder score', data=regional_happiness, color='lightblue')
plt.title('Average Happiness Score by Region')
plt.xlabel('Region')
plt.ylabel('Average Happiness Score (Ladder score)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()
plt.savefig('happiness_by_region.png')
plt.show()

print(regional_happiness)
plt.savefig('happiness_by_region_table.png')
```

In conclusion, the highest average happiness by region overall is in North America while the least happy region is sub-saharan Africa. the top ten countries that rank the highest are found in Europe while the top ten countries that rank the lowest are found in sub-saharan Africa and South Asia. Factors that may contribute to such disparities have to do with lower GDP: When we compare their GDP to their happiness, countries that have the lowest ones tend to score lower on the happiness scale, other factors that contribute to their suffering have to do with poor standard of health, which makes it hard for people to live with healthy standards such as having access to advanced medicine, and clean water in order to survive longer. In addition to that, with little to no social support, it is hard for citizens of those nations to be able to thrive in the way they should be able to. Another major factor would also be the freedom to make life choices. If a citizen is not even allowed to exercise their human rights, it could be hard for them to have a good life. As such, there would also be little opportunity offered to them throughout their lives, and they would not be able to survive in a well-cared for environment. It would also be hard for a citizen to open a new business if they were not allowed to do that, letting the status quo remain the same since new business opportunities could be denied.

What this data analysis shows us is that even though a country's GDP is important to make its citizens happy and feeling well-cared for, other factors are just as important to consider.
