```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df =  pd.read_csv('crime.csv')
df.head()
```

```python
#Task1: Basic bar chart for crime rates in 2023
plt.bar(df['offense'], df['Offenses per 100000 people(2023)'], color= 'red')
plt.xlabel('Offense')
plt.xticks(fontsize=6)
plt.ylabel('Offenses per 100000 people(2023)')
plt.title('Offenses rate per 100000 people(2023)')
plt.grid(True)
plt.show()
```

```python
#Task 2: Change in crime rates(2022-2023) with unique colors
colors = ['red', 'green', 'blue', 'purple', 'orange', 'brown']
plt.barh(df['offense'], df['Change in offenses per 100000 people (2022-2023)'], color= colors)
plt.xlabel('Change in offenses per 100000 people (2022-2023)')
plt.ylabel('Offense')
plt.title('Change in offenses per 100000 people (2022-2023)')
plt.show()
```

1. Why does this bar graph make more sense than Task 1's bar graph?

  The first graph lists the data for the crimes committed during the year 2023, whereas the second graph gives you a clear idea of how high or low those crimes have occurred as a comparison between two years. There is more information in the second graph since it is a comparison and gives us the percentage value of the offenses.

```python
#Task 2: Change in crime rates(2022-2023) with unique colors
colors = ['red', 'yellow', 'maroon', 'turquoise', 'orange', 'brown']
plt.bar(df['offense'], df['Change in offenses per 100000 people (2022-2023)'], color= colors)
plt.xlabel('Change in offenses per 100,000 people (2022-2023)')
plt.ylabel('Offense')
plt.xticks(fontsize=6, rotation=45)
plt.title('Percentage of change in offenses between 2022 and 2023')
plt.grid(True)
plt.show()
```

```python
#Task 2: Change in crime rates(2022-2023) with unique colors
colors = ['violet', 'red', 'maroon', 'turquoise', 'orange', 'teal']
plt.barh(df['offense'], df['Change in offenses per 100000 people (2022-2023)'], color= colors)
plt.xlabel('Offenses')
plt.ylabel('Change in offenses per 100000 people (2022-2023)')
plt.xticks(fontsize=6)
plt.title('Data Showing how much lower crime was in 2023 compared to 2022')
plt.grid(which='major', linewidth=0.5, linestyle='-')
plt.grid(which='minor', linestyle='-')
plt.minorticks_on()
plt.show()
```
