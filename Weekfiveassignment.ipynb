```python
#2: Fare Distribution
sns.histplot(titanic['fare'], kde=True) #how many tickets were purchased at what prices.
plt.show()
```

```python
sns.kdeplot(titanic['fare'], fill=True) #how many tickets were purchased at what prices with a kde plot.
plt.show()
```

The histograms and boxplot tells us that although there were more tickets bought at cheaper prices, the outliers represent those tickets that were bought at a higher price, suggesting that those belonged to  passengers in first class.
```python
#3: Fare and survival rates
plt.figure(figsize=(10, 6))
sns.kdeplot(data=titanic, x='fare', hue='survived', fill=True)
plt.show()  #kde plot of the number of people who survived according to their socioeconomic status.
```

As we can see, most people who did not survive are those whose tickets cost between 0-100, with some outliers going into the 250's. This suggest that passengers in the third class saw the most fatalities, while those in different classes fared better. The classes were clearly divided, and some were favored over others. This does not come as any surprise since it is clear that those who paid more were better accomodated, and even got rescued faster.

```python
#5. Age of survivors and non-survivors
plt.figure(figsize=(10, 6))
sns.kdeplot(data=titanic, x='age', hue='survived', fill=True)
plt.show() #kde plot comparing age distribution among survivors and non-survivors.
```

This graph shows us that the majority of non-survivors were between the ages of 20-30 years old, while children had the highest survival rate, followed by older adults. This suggests that children and the elderly were prioritized in the lifeboats. Usually people(in those days mostly men) of 'working age' would set out to do tasks that facilitates saving people when a tragedy strikes; making sure that women, children, and the elderly are taken care of before taking care of themselves.

```python
#5: Pyramid histograms for age survival by sex
def create_population_pyramid(data, title):
    """Creates a vertical population pyramid with age and sex, split by survival status."""

    survived = data[data['survived'] == 1]
    not_survived = data[data['survived'] == 0]

    fig, axes = plt.subplots(ncols=2, sharey=True, figsize=(10, 6))

    # Survived
    sns.histplot(x=survived[survived['sex'] == 'male']['age'],
                 ax=axes[0], color='skyblue', label='Male', bins=20, kde=True)
    sns.histplot(x=survived[survived['sex'] == 'female']['age'],
                 ax=axes[0], color='lightcoral', label='Female', bins=20, kde=True)
    axes[0].set_title('Survived')
    axes[0].set_xlabel('Age')
    axes[0].legend()

    # Not Survived
    sns.histplot(x=not_survived[not_survived['sex'] == 'male']['age'],
                 ax=axes[1], color='skyblue', label='Male', bins=20, kde=True)
    sns.histplot(x=not_survived[not_survived['sex'] == 'female']['age'],
                 ax=axes[1], color='lightcoral', label='Female', bins=20, kde=True)
    axes[1].set_title('Not Survived')
    axes[1].set_xlabel('Age')
    axes[1].legend()

    fig.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()

create_population_pyramid(titanic, 'Titanic Passengers: Age and Sex by Survival')
```

```python
#7: Bar graph, line plot, and scatterplot

# 1. Bar Graph: Survival Rate by Passenger Class
survival_rates_by_class = titanic.groupby('pclass')['survived'].mean()
plt.figure(figsize=(8, 5))
sns.barplot(x=survival_rates_by_class.index, y=survival_rates_by_class.values)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# 2. Line Plot: Average Fare by Age
average_fare_by_age = titanic.groupby('age')['fare'].mean()
plt.figure(figsize=(10, 6))
sns.lineplot(x=average_fare_by_age.index, y=average_fare_by_age.values)
plt.title('Average Fare by Age')
plt.xlabel('Age')
plt.ylabel('Average Fare')
plt.show()

# 3. Scatter Plot: Age vs. Fare, Colored by Survival
plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived')
plt.title('Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

These three graphs confirm the findings we did earlier in this assignment. In the bar graph, we are told that the highest number of people who survived were in first class, the second being second, and third last, which makes sense according to their priority. The line plot gives you the average fare by age, denoting that the majority of tickets were held by infants, then people in their mid-thirties, then those in their late 50's, and finally those in their mid-sixties. The scatterplot tells you that children under 10, some people in their 20's, those in their 40's to their 60's had the highest number of survivors, with one outlier in their 80's and a select few in their mid-thirties(who are also outliers in that they paid the highest fares, which would explain why those passengers survived, given their VIP status).

**Conclusion**: To survive the Titanic, one needed to have money, which would afford them a higher passenger class, and first priority if something goes wrong. In addition to that, women and children were also given priority since they had to be protected according to the laws of chivalry.
