```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#style
#plt.style.use('seaborn-whitegrid')
plt.style.use('ggplot')
```

```python
plt.style.available #this gives you all available styles
```

```python
x=np.linspace(0,2*3.14,100)
plt.scatter(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dotted')
```

```python
plt.plot(x, x+4, linestyle='--', color='g')
```

```python
plt.plot(x, x+4, '--g')
plt.plot(x, x+3, '--r')
plt.xlim(-2,5)
```

```python
plt.plot(x, x+0, '--b')
plt.xlim(-2, 15)
plt.ylim(-2, 15)
plt.axis('tight')
```

```python
plt.plot(x, x)
plt.xlim(0, 10)
plt.ylim(0, 2)
```

```python
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.title('sine curve')
plt.xlabel("x")
plt.ylabel("sin(x) and cos(x)")
plt.legend()
plt.xlim(-1.5, 1.5)
plt.figure(figsize=(20,10))
```
