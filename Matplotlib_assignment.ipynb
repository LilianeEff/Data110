```python
#1. Setup and Styles
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use( 'seaborn-v0_8-whitegrid')
plt.style.use('ggplot')
```

```python
#2. Basic Line Plot
def f(x):
  return x**2 - 2*x + 1

x = np.linspace(-2, 4, 100)
y = f(x)
plt.plot(x, y, label='f(x) = x^2 - 2x + 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Quadratic Formula')
plt.legend()
plt.grid()
plt.xlim(-2, 4)
plt.ylim(0, 10)
plt.show()
```

```python
#3. Multiple lines
x = np.linspace(-2, 2, 100)
x2 = np.linspace(0.1, 2, 100)
y1 = x**3
y2 = np.exp(x)
y3 = np.log(x)
plt.plot(x, y1, label='f(x) = x³', color='blue')
plt.plot(x, y2, label='g(x) = eˣ', color='red')
plt.plot(x2, y3, label='h(x) = lnˣ', color='orange')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Three Different Plots")
plt.legend()
plt.grid()
plt.xlim(-2, 4)
plt.ylim(0, 10)
plt.show()
```

```python
#4. Styling Lines
x= np.linspace(-2, 2, 100)
y= x**2
y1= x**3
y2= x**4
y3= x**5
plt.plot(x, y, label='f(x) = x²', color='blue', linestyle='--')
plt.plot(x, y1, label='g(x) = x³', color='red', linestyle='-.')
plt.plot(x, y2, label='h(x) = x⁴', color='green', linestyle=':')
plt.plot(x, y3, label='i(x) = x⁵', color='orange', linestyle='-')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Five Different Plots")
plt.legend()
plt.grid()
plt.xlim(-2, 4)
plt.ylim(0, 10)
plt.show()
```

```python
# 5. Axes Limits and Scaling
x = np.linspace(-10, 10, 400)
y = 1/(x**2 + 1)

#custom limits
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.plot(x, y)
plt.title('f(x) = 1/(x^2 + 1)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim(-10, 10)
plt.ylim(0, 1.1)
plt.grid(True)

#Reversed Axes
plt.subplot(1, 3, 2)
plt.plot(y, x)
plt.title('Reversed Axes')
plt.xlabel('f(x)')
plt.ylabel('x')
plt.grid(True)

# tight axis
plt.subplot(1, 3, 3)
plt.plot(x, y)
plt.title('plt.axis(\'tight\')')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axis('tight')
plt.grid(True)
plt.tight_layout()

plt.show()

#  'equal' axis
plt.figure()
plt.plot(x, y)
plt.title('plt.axis(\'equal\')')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
#6. Labels and Legends
x= np.linspace(0, 5, 100)
y= np.sqrt(x)
y1= x**(1/3)
plt.plot(x, y, label='f(x) = √x')
plt.plot(x, y1, label='g(x) = x^(1/3)')
plt.title('Square Root and Cube Root')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
#7. Basic Scatter Plot
x= np.random.rand(50)
y= np.random.rand(50)
plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.grid(True)
plt.show()
```

```python
#8. Customizing Scatter Plots
x = np.linspace(0, 4*np.pi, 100) # Define x before calculating y
y= x * np.sin(x)
plt.scatter(x, y, s=60, marker='o', linewidths=2,
            facecolor='lightblue', edgecolor='blue', label='f(x) = x*sin(x)')
plt.plot(x, y, linewidth=1.5, color='green')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x*sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#11. Style Comparison
y= x * np.cos(x)
x_line = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y_line = f(x_line)
x_scatter = np.arange(-2 * np.pi, 2 * np.pi, np.pi/4)
y_scatter = f(x_scatter)

# --- Default Style ---
plt.figure(figsize=(8, 6))
plt.plot(x_line, y_line, label='f(x) = x*cos(x)')
plt.scatter(x_scatter, y_scatter, label='Scatter Points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x*cos(x) (Default Style)')
plt.grid(True)
plt.legend()
plt.show()

# --- Seaborn-whitegrid Style ---
plt.figure(figsize=(8, 6))
plt.style.use('seaborn-whitegrid')
plt.plot(x_line, y_line, label='f(x) = x*cos(x)')
plt.scatter(x_scatter, y_scatter, label='Scatter Points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x*cos(x) (Seaborn-whitegrid Style)')
plt.legend()
plt.show()

# --- Ggplot Style ---
plt.figure(figsize=(8, 6))
plt.style.use('ggplot')
plt.plot(x_line, y_line, label='f(x) = x*cos(x)')
plt.scatter(x_scatter, y_scatter, label='Scatter Points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x*cos(x) (Ggplot Style)')
plt.legend()
plt.show()

# --- Comparison ---
# Default Style:
# - Classic Matplotlib look with a white background and blue line/markers.
# - Grid lines are present by default.
# - Simple and familiar appearance.

# Seaborn-whitegrid Style:
# - Clean and modern aesthetic with a white background and subtle gray grid lines.
# - Emphasizes the data points and lines.

# Ggplot Style:
# - Gray background with white grid lines and distinct color palette.
# - Visually appealing and can enhance data visualization.

# Most appropriate style:
# - For this specific plot, the 'seaborn-whitegrid' style might be most appropriate.
# - It presents the data clearly and concisely without unnecessary distractions.
# - The gray grid lines provide a subtle way to increase readability for any viewer.
```
