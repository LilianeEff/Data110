```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import folium
from folium.plugins import HeatMap
import seaborn as sns
```

```python
df=pd.read_csv('/content/drive/MyDrive/BPD_Arrests.csv')
df.head()
```

```python
cleaned_df = df.dropna()
cleaned_df.head()
```

```python

cleaned_df = df.dropna(subset=['Latitude', 'Longitude'])
cleaned_df.head()
```

####*- Create a combined heatmap visualization for the **BPD_Arrests.csv** dataset, with at least two interactive layers based on **Gender** (e.g., one layer for arrests involving males and another for arrests involving females).*

```python
gender_location_counts = cleaned_df.groupby(['Gender', 'Latitude', 'Longitude']).size().reset_index(name='IncidentNumber')

print(gender_location_counts.head())
```

```python
# Separate data by gender
male_counts = gender_location_counts[gender_location_counts['Gender'] == 'M']
female_counts = gender_location_counts[gender_location_counts['Gender'] == 'F']

# Convert to lists for HeatMap
male_heatmap_data = male_counts[['Latitude', 'Longitude', 'IncidentNumber']].values.tolist()
female_heatmap_data = female_counts[['Latitude', 'Longitude', 'IncidentNumber']].values.tolist()
```

```python
m = folium.Map(location=[cleaned_df['Latitude'].mean(), cleaned_df['Longitude'].mean()], zoom_start=12)

# Male HeatMap
if len(male_heatmap_data) > 0:
    male_heatmap = HeatMap(male_heatmap_data, name="Male Arrests", gradient={0.2: 'blue', 0.5: 'cyan', 0.9: 'lime'})
    m.add_child(male_heatmap)

# Female HeatMap
if len(female_heatmap_data) > 0:
    female_heatmap = HeatMap(female_heatmap_data, name="Female Arrests", gradient={0.2: 'red', 0.5: 'orange', 0.9: 'yellow'})
    m.add_child(female_heatmap)

folium.LayerControl().add_to(m)
m.save("arrests_heatmap_by_gender.html")
m
```
