```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/LilianeEff/Data110-22016/refs/heads/main/dataset/world_countries.csv')
```

```python
df.head()
```

```python
mapping_data= df.rename(columns={'country':'Country', 'GDP':'GDP_per_capita', 'population':'Population'})
fig=px.choropleth(
    mapping_data,
    locations='Country',
    locationmode='country names',
    color='GDP_per_capita',
    hover_name='Country',
    title= "choropleth map of GDP per country",
    color_continuous_scale=px.colors.sequential.Viridis
)
fig.show()
```

```python
mapping_data= df.rename(columns={'country':'Country', 'GDP':'GDP_per_capita', 'population':'Population'})
fig=px.choropleth(
    mapping_data,
    locations='Country',
    locationmode='country names',
    color='GDP_per_capita',
    hover_name='Country',
    title= "choropleth map of GDP per country",
    color_continuous_scale=px.colors.sequential.Viridis
)
fig.show()
```

```python
gdp_by_region=df.groupby('Region')['GDP'].sum(). index
gdp_by_region = df.groupby('Region')['GDP'].sum().reset_index()
fig = px.bar(gdp_by_region, x='Region', y='GDP', title='GDP by Region')
fig.show()
```

```python
plt.barh(df['Region'], df['GDP'])
plt.xlabel('GDP')
plt.ylabel('Country')
plt.title('GDP by Region')
plt.show()
```

```python
population_by_region = df.groupby('Region')['Population'].sum().reset_index()
population_by_region = population_by_region.sort_values(by='Population', ascending=False)
plt.barh(population_by_region['Region'], population_by_region['Population'])
plt.xlabel('Region')
plt.ylabel('Population')
plt.title('Population by Region')
plt.show()
```

```python
treemap_data= df[['Country', 'Population']].dropna()
treemap_data = treemap_data.sort_values(by='Population', ascending=False)
treemap_data.info()
```

```python
treemap_data['Population'] = treemap_data['Population'] / 1e6
treemap_data = treemap_data.nlargest(14, 'Population')
treemap_data = df.groupby('Region')['Population'].sum().reset_index()
treemap_data = treemap_data.sort_values(by='Population', ascending=False)
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
squarify.plot(sizes=treemap_data['Population'], label=treemap_data['Region'], color=colors, pad=True)
plt.show()
```
